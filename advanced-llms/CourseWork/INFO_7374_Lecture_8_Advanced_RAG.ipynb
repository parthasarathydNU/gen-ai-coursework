#### Query Rewriting / Query Expansion / Sub Queries

![](https://i.imgur.com/mhPZTUD.png)


Query Rewriting for Retrieval-Augmented Large Language Models

https://arxiv.org/pdf/2305.14283.pdf


**No training**
Ask an LLM to rewrite queries to search for information that is needed.

**With Training**
A small LLM is trained to rewrite queries.

1. Create a pseudo dataset for query rewriting task for warm up.
    
    i. Prompt the LM to rewrite original questions $x$ in the training set and collect the generated samples $\tilde{x}$ as pseudo labels.
    
    ii.  Filter pseudo labels. Those that get correct predictions from the LLM are selected to be a part of warm up dataset.
2. Fine tune Query Rewriter with RLHF
    i. Use Reward model to evaluate the quality of results
    ii. Query rewriter (small LM) is tuned to increase probabilties of tokens that led to higher rewwards, and decrease probabilities of tokens that led to lower rewards.
    
    Note that the LLM that generates the final response is frozen.

https://arxiv.org/abs/2212.10496

Hyde is a retrieval method that stands for Hypothetical Document Embeddings (HyDE). It is a method used to enhance retrieval by generating a hypothetical document for an incoming query.

The document is then embedded, and that embedding is utilized to look up real documents that are similar to the hypothetical document.

The underlying concept is that the hypothetical document may be closer in the embedding space than the query.

#### How to Pick an Embedding Model

Things to consider:

1. Embedding dimension size: The dimensionality of the embeddings can impact the performance of the RAG system. Reducing the dimensionality might lead to information loss, but it can help focus on the most important aspects of the data.
2. Retrieval Accuracy: The retrieval mechanism is a key part of a successful RAG application. The embedding model should be capable of generating embeddings that enable accurate and efficient retrieval of relevant documents.
3. Model size: The choice of embedding model can also be influenced by the available computational resources and infrastructure. Some models might require more computational power or storage than others, so it's important to consider these factors when choosing a model.

![](https://imgur.com/u44SwOp.png)

4. Context length: The context length of the data that embedding model was trained with influences the granularity of the text that is embedding. If the application requires very nuanced information, perhaps its is better to chunk the information into smaller components and use an embedding model that is more specialized for smaller contexts. Additionally, you want to make sure that the context length of the text that you are embedding follows that of the training distribution.

5. Application: Different embedding models are better at different tasks, it may be useful to put a model that performs best at a task most similar to your application.

![](https://i.imgur.com/6obLw8z.png)

#### Bootstrapping Data for Training/Finetuning an Embedding Model

![](https://production-media.paperswithcode.com/tasks/34886962-1da9-4e2d-a833-40699515230a.png)

Recall that for training embedding models using the InfoNCE objective, we typically need two components:

1. positive pairs: pairs of text samples that are semantically related or similar. These can be sentences, paragraphs, or documents depending on the granularity of the embeddings you want to learn.
* Examples of positive pairs include:
    * Sentences from the same document or paragraph.
    * Question-answer pairs from a QA dataset.
    * Parallel sentences from bilingual corpora in machine translation.
2. Negative Pairs: pairs of text samples that are unrelated or dissimilar
* Negative samples include:
    * Random sampling from the dataset, ensuring that the negative sample is not related to the positive anchor.
    * Hard negative mining, where the negative samples are chosen to be semantically close but not identical to the positive anchor, making the contrastive task more challenging.

# Maximum Inner Product Search (MIPS)

- **Inner Product (Dot Product):** The inner product of two vectors $\mathbf{a}$ and $\mathbf{b}$ is a scalar value obtained by multiplying corresponding elements of the vectors and then summing those products. For vectors $\mathbf{a}$ and $\mathbf{b}$, the inner product is denoted as $\mathbf{a} \cdot \mathbf{b}$ or $\langle \mathbf{a}, \mathbf{b} \rangle$.

- **Cosine Similarity:** Cosine similarity is a measure of similarity between two non-zero vectors based on the cosine of the angle between them. It is calculated as the inner product of the vectors divided by the product of their magnitudes (or Euclidean norms).


Main difference between cosine similarity and inner product is that cosine similarity normalizes by the product of the norm of $a$ and $b$

```python
from transformers import AutoTokenizer, AutoModel
import torch
import faiss
import numpy as np

# Initialize the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

# List of strings from different topics
strings = [
    "The quick brown fox jumps over the lazy dog.",
    "Climate change is a global challenge that affects everyone.",
    "Python is a popular programming language for data science.",
    "The stock market can be volatile and unpredictable.",
    "Artificial intelligence is transforming industries.",
    "Healthy eating is essential for maintaining a balanced diet.",
    "The solar system consists of the sun and its orbiting planets.",
    "Music has the power to evoke emotions and memories.",
    "Exercise is important for physical and mental health.",
    "Quantum computing is an emerging field with potential."
]

# Encode the strings into embeddings
def encode(texts):
    with torch.no_grad():
        encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
        output = model(**encoded_input)
        return output.last_hidden_state.mean(dim=1).numpy()

embeddings = encode(strings)

# Create an IVF index with PQ quantization
dimension = embeddings.shape[1]
nlist = 4  # Number of clusters
m = 2      # Number of PQ subvectors
bits = 2   # Number of bits for quantization
quantizer = faiss.IndexFlatL2(dimension)
index = faiss.IndexIVFPQ(quantizer, dimension, nlist, m, bits)
index.train(embeddings) # training involves clustering the vectors into partitions
index.add(embeddings)

# Example query: Find the 3 most similar sentences to the first one in the list
query_embedding = encode([strings[0]])
distances, indices = index.search(query_embedding, k=3)

# Print the results
print("Query:", strings[0])
for i, idx in enumerate(indices[0]):
    print(f"Rank {i+1}: {strings[idx]}")
```

# Evaluation of RAG Systems

## RAGAS: Automated Evaluation of Retrieval Augmented Generation

https://arxiv.org/abs/2309.15217

Use LLMs to evaluate answer faithfulness, answer relevance, context relevance

#### Answer Faithfulness

Answer should be grounded in the given context

An answer is faithful to the context if the claims made in the answer can be inferred from the context.



1. Extract a set of statements. Goal is to decompose longer sentences into shorter and more focused assertions

*Given a question and answer, create or or more statements from each sentence in the given answer*

*question: [question]*

*answer: [answer]*

2. Then, use another query to determine if each statement can be inferred from the context

*Consider the given context and following statements, then determine whether they are supported by the information present in the context. Provide a brief explanation for each statement before arriving at the verdict (Yes/No). Provide a final verdict for each statement in order at the end in the given format. Do not deviate fro mthe specified format.*

*statement: [statement 1]*

...


*statement: [statement n]*

3. Final faithfulness score is proportion of statements that were supported according to the LLM.





#### Answer Relevance

Answer should address the *question*

Answer is relevant if it directly addresses the question in an appropriate way. Does not take into account factuality, but penalizes cases where the answer is incomplete or where it contains redundant information. To estimate answer relevance, for the given answer, we prompt the LLM to generate $n$ potenntial questions based on the answer.

*Generate a question for the given answer*
*answer: [answer]*

Then, embed all the questions using an embedding model. For each question, compute the cosine similarity between the generated question and the original question.

The answer relevance is the average similarity across the $n$ generated questions.

#### Context Relevance

Retrieved context should be focused, contain little irrelevant information. This is important because of the cost associated with long context and when the context is too long, the LM can not effectively use the middle of the context.





Context is relevant to the extent that it exclusively contains information that is needed to answer the question. In particular, this metric aims to penalize the inclusion of redundant information.


*Please extract relevant sentences from the provided context that can potentially help answer the following question. If not relevant sentences are found, or if you believe the question cannot be answered from the given context, return the phrase "Insufficient Information". While extracting candidate sentences you're not allowed to make any changes to senteces from given context.*


 $$CR=\frac{\text{# of extracted sentences}}{\text{# of sentences in context}}$$
